```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
gumbel = sts.gumbel_l(0, 1) #распределение Гумбеля
sample = gumbel.rvs(1000) #генерация выборки для построения гистограммы
```

Распределение Гумбеля используется для моделирования распредений максимумов или минимумов других различных распределений. <br>
Плотность вероятности равна $e^{-(e^{x}-x)}$. Распределение похоже на нормальное, но со смещенным влево центром. <br>
Мат. ожидание равно $-\gamma$, где $\gamma$ - постоянная Эйлера-Маскерони, численно равная примерно 0.577. <br>
Дисперсия равна $\zeta(2)$ или $\frac{\pi^2}{6}$.

```python
#построение гистограммы и соответсвущего ей графика функции распределения
x = np.linspace(-7,3,100)
pdf = gumbel.pdf(x)
plt.hist(sample, bins=40, density=True)
plt.plot(x, pdf, label='Gumbel distribution PDF')
plt.legend()
plt.ylabel('$f(x)$ and fraction of samples')
plt.xlabel('$x$')
```

```python
#генерируем по тысяче выборок объемов 5, 10 и 50. Дальше - записываем в массив средние этих выборок.
means_5 = []
for i in range(1000):
    means_5.append(gumbel.rvs(5).mean())
means_10 = []
for i in range(1000):
    means_10.append(gumbel.rvs(10).mean())
means_50 = []
for i in range(1000):
    means_50.append(gumbel.rvs(50).mean())
```

```python
#Строим гистограмму распределения средних на выборках из пяти элементов и сравниваем ее с нормальным распределением
#с соответсвующими мат. ожиданием и дисперсией
norm_5 = sts.norm(-np.euler_gamma, np.pi/np.sqrt(6*5))
x = np.linspace(-3,1.3,100)
pdf = norm_5.pdf(x)
plt.plot(x, pdf, label='$N(-\gamma, \\frac{\pi^2}{6\\cdot5})$')
plt.legend(loc='upper left', fontsize = 'x-large')
plt.hist(means_5, bins=40, density=True)
plt.ylabel('fraction of means')
plt.xlabel('$mean$')
```

```python
#Строим гистограмму распределения средних на выборках из десяти элементов и сравниваем ее с нормальным распределением
#с соответсвующими мат. ожиданием и дисперсией
norm_10 = sts.norm(-np.euler_gamma, np.pi/np.sqrt(6*10))
x = np.linspace(-2.6,1,100)
pdf = norm_10.pdf(x)
plt.plot(x, pdf, label='$N(-\gamma, \\frac{\pi^2}{6\\cdot10})$')
plt.legend(loc='upper left', fontsize = 'x-large')
plt.hist(means_10, bins=40, density=True)
plt.ylabel('fraction of means')
plt.xlabel('$mean$')
```

```python
#Строим гистограмму распределения средних на выборках из пятидесяти элементов и сравниваем ее с нормальным распределением
#с соответсвующими мат. ожиданием и дисперсией
norm_50 = sts.norm(-np.euler_gamma, np.pi/np.sqrt(6*50))
x = np.linspace(-1.2,0.1,100)
pdf = norm_50.pdf(x)
plt.plot(x, pdf, label='$N(-\gamma, \\frac{\pi^2}{6\\cdot50})$')
plt.legend(loc='upper right', fontsize = 'x-large')
plt.hist(means_50, bins=40, density=True)
plt.ylabel('fraction of means')
plt.xlabel('$mean$')
```

Из приведенных графиков видно, что в соответсвие с предсказаниями ЦПТ распределение средних в выборках хорошо описывается
нормальным распределением. Причем разброс уменьшается с примерно $4$ до примерное $1$, что также вполне согласуется 
с нормальным распределением, т.к. область $99.7$-процентной вероятности изменяется с $\sim3.44$ до $\sim1.09$.
