##                            ЦПТ для Экспоненциального распределения

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import expon
from scipy.stats import norm
import math
%matplotlib inline
```

Ссылка на теоретические сведения об экспоненциальном распределении https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D1%81%D0%BF%D0%BE%D0%BD%D0%B5%D0%BD%D1%86%D0%B8%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5

Функция распределения      f(x) = 1-e^{-\lambda x}
Математическое ожидание    EX = lambda^-1
Дисперсия                  DX = lambda^-2

```python
r = expon.rvs(size=1000)                     # генерируем выборку случайных величин размера 1000 
x = np.linspace(-0.5, 4,1000)                # генерируем массив значений x
plt.plot(x, expon.pdf(x), color='r')         # рисуем график теоретической плотности распределения случайной величины
plt.hist(r, 100, density=True, histtype='stepfilled')              # гистограмма выборки
plt.xlabel('X')
plt.ylabel('1-e^{-lambda x}')
plt.show()
```

```python
lb = 1.              # параметр лямбда
EX = 1./lb           # математическое ожидание
DX = 1./lb**2        # Дисперсия
```

```python
n = 5                                  # в этом блоке мы генерируем 1000 выборок(выборочных средних) объема n=5
n5 = []
for i in range(1000):
    rvs = expon.rvs(size=n)            # генерируем выборки объема n=5
    n5.append(np.sum(rvs)/n)           # высчитываем выборочное среднее и записываем в массив 
```

```python
plt.hist(n5, 100, density=True, histtype='stepfilled')       # создаем гистограмму выборочных средних
plt.plot(x, norm.pdf(x, EX, (math.sqrt(DX/n))))              # рисуем график плотности нормального распределения
''' т.к. функция принимает на вход стандартное отклонение, расчитываем его значение
для каждого DX и n'''
plt.xlabel('X')
plt.ylabel('')
plt.show()
```

```python
n = 10                                # в этом блоке мы генерируем 1000 выборок(выборочных средних) объема n=10
n10 = []
for i in range(1000):
    rvs = expon.rvs(size=n)           # генерируем выборки объема n=10
    n10.append(np.sum(rvs)/n)         # высчитываем выборочное среднее и записываем в массив 
```

```python
plt.hist(n10, 100, density=True, histtype='stepfilled')      # создаем гистограмму выборочных средних
plt.plot(x, norm.pdf(x, EX, (math.sqrt(DX/n))))              # рисуем график плотности нормального распределения
''' т.к. функция принимает на вход стандартное отклонение, расчитываем его значение
для каждого DX и n'''
plt.xlabel('X')
plt.ylabel('')
plt.show()
```

```python
n = 50                                            # в этом блоке мы генерируем 1000 выборок(выборочных средних) объема n=10
n50 = []
for i in range(1000):
    rvs = expon.rvs(size=n)                       # генерируем выборки объема n=10
    n50.append(np.sum(rvs)/n)                     # высчитываем выборочное среднее и записываем в массив 
```

```python
plt.hist(n50, 100, density=True, histtype='stepfilled')     # создаем гистограмму выборочных средних
plt.plot(x, norm.pdf(x, EX, (math.sqrt(DX/n))))             # рисуем график плотности нормального распределения
''' т.к. функция принимает на вход стандартное отклонение, расчитываем его значение
для каждого DX и n'''
plt.xlabel('X')
plt.ylabel('')
plt.show()
```

# Вывод

Проведя наблюдения за тем как гистограмма выборочных средних и график плотности нормального распределения сходятся при росте значения n, можно сделать вывод, что распределение выборочных средних достаточно хорошо описывается нормальным распределением, что и является утверждением центральной предельной теоремы. С ростом n точность нормальной аппроксимации увеличивается.
